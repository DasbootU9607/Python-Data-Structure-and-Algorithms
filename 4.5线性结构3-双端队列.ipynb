## 3.双端队列

### 3.1 双端队列的ADT及实现

```python
class Deque:
    def __init__(self):
        self.items=[]
    def isEmpty(self):
        return self.items==[]
    def addFront(self,item):
        self.items.append(item)
    def addRear(self,item):
        self.items.insert(0,item)
    def removeFront(self):
        return self.items.pop()
    def removeRear(self):
        return self.items.pop(0)
    def size(self):
        return len(self.items)
```

### 3.2 双端队列的应用

#### （1）回文词判定

```python
def palindrome_check(word):
    dq=Deque()
    for ch in word:
        dq.addRear(ch)
    stillequal=True
    while dq.size()>1 and stillequal:
        if dq.removeFront()!=dq.removeRear():
            stillequal=False
    return stillequal

print(palindrome_check("radar"))
print(palindrome_check("happy"))

```
