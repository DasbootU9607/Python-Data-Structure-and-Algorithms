### 2.2.1变位词判断

```python
# 问题：检查两个词是否存在字母换位关系
#解法一：逐字检查
def anagram_soltion1(s1,s2):
    list2=list(s2)
    if s1==s2 or len(s1)!=len(s2):
        return False
    else:
        for i in range(len(s1)):
            found=False
            for j in range(len(list2)):
                if s1[i]==list2[j]:
                    found=True
                    list2[j]=None
            if not found:
                return False
        for char in list2:
            if char is not None:
                return True
#解法2：排序比较
def anagram_solution2(s1,s2):
    list1=list(s1)
    list2=list(s2)
    list1_sort=sorted(list1)
    list2_sort=sorted(list2)
    if list1==list2 or len(list1)!=list2:
        return False
    else:
        return list1_sort==list2_sort

#解法3：计数比较
def anagram_solution3(s1,s2):
    if len(s1) != len(s2) or s1 == s2:
        return False
    counter=[0]*26
    for char in s1:
        index=ord(char)-ord('a')
        counter[index]+=1
    for char in s2:
        index=ord(char)-ord('a')
        counter[index]-=1
    return counter==[0]*26
```
